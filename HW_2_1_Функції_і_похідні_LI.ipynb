1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
# Імпортуємо бібліотекі
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Визначаємо символичну змінну
x = sp.symbols('x')

# Функція
f = 10*x - 5
# Похідна
df = sp.diff(f, x)
print('Похідна функції ', df)
```

```python
# Задаємо точки
xs = np.linspace(0, 10, 100)
f_xs = 10*xs - 5
df_xs = np.linspace(10, 10, 100)
#print(df_xs)

# Будуємо графіки фукції та похідної
plt.figure(figsize=(10, 5))
plt.plot(xs, f_xs, color='blue', label="f(x) = 10*x - 5")
plt.plot(xs, df_xs, color='green', label="f'(x) = 10")
plt.legend();
```

Відповідь 1.
Із графіка функціі ми бачимо, що збільшення кількості годин веде до збільшення оцінки, але чисто технично оцінювання починається с 0,5 години. А із графіка похідної робимо висновок, що кожну годину навчання оцінка зростае на 10 одиниць, швідкість зростання оцінки не збільшується.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
# Задаємо функції та символичну змінну
x = sp.symbols('x')
f1 = sp.sin(x)
f2 = sp.ln(x)

# Обчисляємо для них похідні
df1 = sp.diff(f1, x)
df2 = sp.diff(f2, x)
print('Похідна функції sin(x): ', df1)
print('Похідна функції ln(x): ', df2)
```

```python
xs = np.linspace(1, 10, 20)
plt.figure(figsize=(10,5))
# Будуємо графіки функцій
plt.plot(np.sin(xs), xs, 'r', label='f1(x) = sin(x)')
plt.plot(np.log(xs), xs, 'b', label='f2(x) = ln(x)')

# будуємо графіки похідних функцій
plt.plot(np.cos(xs), xs, 'r--', label="f1'(x) = cos(x)")
plt.plot(1 / xs, xs, 'b--', label="f2'(x) = 1/x")
plt.legend();
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
f = x**2 - 5*x + 1
df = sp.diff(f, x)
print('Похідна функції загальних витат виробництва:', df)
```

```python
xs = np.linspace(0, 100, 100)
f_xs = xs**2 - 5*xs + 1
df_xs = 2*xs - 5
plt.figure(figsize=(10,5))
plt.plot(f_xs, xs, 'g', label='Загальні витрати компанії')
plt.plot(df_xs, xs, 'g--', label='Швидкість зростання загальних витрат компанії')
plt.legend();
```
